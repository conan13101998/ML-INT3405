```python
import numpy as np
import pandas as pd
import datetime

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

plt.style.use('seaborn-darkgrid')
plt.rc('figure', figsize=(16,10))
plt.rc('lines', markersize=4)

import yfinance as yf

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.stats.stats import pearsonr
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime.now()
```

```python
def get_close_stock_price(codes, start_date=start, end_date=end):
    data = { code : yf.download(code, start_date, end_date) for code in codes }
    data = { key : pd.DataFrame(value, columns=['Close']).rename(columns={'Close': f'{key}'}) for key, value in data.items() }
    df = None
    for key, value in data.items():
        if df is None:
            df = value
        else:
            df = df.merge(value, left_index=True, right_index=True)
    return df

df = get_close_stock_price(['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'FB'])
df.to_csv('data/close_price_big_5.csv')
```

```python
df = pd.read_csv('data/close_price_big_5.csv', index_col=['Date'], infer_datetime_format=True, parse_dates=['Date'])
df.head()
```

```python
def create_Xy(df, history=1):
    '''df has index as date
    columns are stock price of the companies
    '''
    X, y, dates = [], [], []
    for row_idx in range(history, len(df)):
        y.append( df.iloc[row_idx,:].values.reshape(-1) )
        X.append( df.iloc[row_idx-history:row_idx,:].values.reshape(-1) )
        dates.append(df.index[row_idx])
    X, y = np.array(X), np.array(y)
    return X, y, dates

X,y,dates = create_Xy(df, history=5)
# use 7/10 of the dataset to train (use past price to predict future price)
n_train = len(y) * 7 // 10
X_train, y_train, dates_train, X_test, y_test, dates_test = X[:n_train], y[:n_train], dates[:n_train], X[n_train:], y[n_train:], dates[n_train:]
```

# Bai 2
Investigate if stock price of Big 5 companies are related to each other ( “FAAMG”—Facebook, Amazon, Apple, Microsoft, and Google )

```python
close_stock_price_arr = []
for idx in df.columns:
    close_stock_price_arr.append(np.array(df[idx]))
```

```python
corr_matrix = np.zeros((5,5))

for i in range(len(close_stock_price_arr)):
    for j in range(len(close_stock_price_arr)):
        corr_matrix[i][j] = pearsonr(close_stock_price_arr[i], close_stock_price_arr[j])[0]
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=np.bool))
f, ax = plt.subplots(figsize=(9, 7))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr_matrix, mask=mask, cmap=cmap, xticklabels=df.columns, yticklabels=df.columns, 
            center=0.0, square=True, cbar_kws={"shrink": .5}, annot=True, vmax=1, vmin=-1, linewidths=1.5)
plt.title("Pearsonr correlation between companies's stock price", fontsize=16)
```

```python
variation_matrix = [[] for i in range(len(df.columns))]

for i in range(len(close_stock_price_arr)):
    for j in range(2, len(close_stock_price_arr[i])):
        variation_matrix[i].append(close_stock_price_arr[i][j]-close_stock_price_arr[i][j-1])
```

```python
plt.figure(figsize=(40,10))
for i in range(len(variation_matrix)):
    plt.plot(variation_matrix[i], label=df.columns[i])
plt.title('Variation of each stock', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Variation', fontsize=14)
plt.legend()
plt.show()
```

```python
def create_Xy(df, history=1):
    '''df has index as date
    columns are stock price of the companies
    '''
    X, y, dates = [], [], []
    for row_idx in range(history, len(df)):
        y.append( df.iloc[row_idx,:].values.reshape(-1) )
        X.append( df.iloc[row_idx-history:row_idx,:].values.reshape(-1) )
        dates.append(df.index[row_idx])
    X, y = np.array(X), np.array(y)
    return X, y, dates

X,y,dates = create_Xy(df, history=5)
n_train = len(y) * 5 // 10
(x_train, y_train, dates_train, x_test, y_test, dates_test) = [X[:n_train],
                                                               y[:n_train],
                                                               dates[:n_train],
                                                               X[n_train:],
                                                               y[n_train:],
                                                               dates[n_train:]]
```

```python
def mean_relative_error(y_pred, y_true):
    return np.mean(np.abs((y_pred - y_true)/y_true))
```

## Cách 1
* Data: Mỗi sample có 25 feature là giá cổ phiếu của Big 5 trong 5 ngày gần nhất
* Model: Sử dụng một Linear Regression cơ bản

```python
clf = LinearRegression()
clf.fit(x_train, y_train)
y_pred_1 = clf.predict(x_test)
```

```python
mre = []
for i in range(len(df.columns)):
    mre.append(mean_relative_error(y_pred_1[:, i], y_test[:, i]))
    print(f'{df.columns[i]}: ', mean_relative_error(y_pred_1[:, i], y_test[:, i]))
print(f'Model mean relative error: {np.mean(mre)}')
```

```python
for code in range(len(df.columns)):
    df_plot = pd.DataFrame({f'{df.columns[code]}_true': y_test[:,code], f'{df.columns[code]}_pred': y_pred_1[:,code]}, index=dates_test)

    # Create subplots to plot graph and control axes
    fig, ax = plt.subplots(figsize=(25,10))
    sns.lineplot(data=df_plot)

    # Format the ticks
    loc = mdates.AutoDateLocator()
    ax.xaxis.set_major_locator(loc)
    ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(loc))

    # Set figure title
    plt.title(f'{df.columns[code]} stock price: True vs. Prediction', fontsize=16)
    # Set x label
    plt.xlabel('Date', fontsize=14)
    # Set y label
    plt.ylabel('Closing Stock Price in $', fontsize=14)

    # Rotate and align the x labels
    fig.autofmt_xdate()

    # Show plot
    plt.show()
```
